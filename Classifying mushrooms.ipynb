```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns: 
    le = LabelEncoder();
    le.fit(df[col].unique());
    df[col] = le.transform(df[col]);
df.head()
```

```python
df = df.sample(frac=1, random_state = 20);

X_train = df.drop(['class'], axis = 1)[0:600]
y_train = df['class'][0:600]
X_test = df.drop(['class'], axis = 1)[6500:8123]
y_test = df['class'][6500:8123]
```

```python
from sklearn.neural_network import MLPClassifier

# train model
nn = MLPClassifier().fit(X_train, y_train);

# evaluate model
print("Accuracy on train-set:", "%.2f" %(nn.score(X_train, y_train)*100), '%')
print("Accuracy on test-set:", "%.2f" %(nn.score(X_test, y_test)*100), '%')
```
